### **Autoregressive (AR) Models Explained in Great Detail**

The **Autoregressive (AR) model** is a widely used statistical approach for analyzing and predicting time series data. The key idea is that the current value of a time series depends linearly on its past values. This makes AR models suitable for data that exhibit autocorrelation, meaning past observations influence future observations.

### **Basic Concept**

An **AR(p)** model predicts the current value of the time series $ Y_t $ as a linear combination of its previous $ p $ values, plus a noise term. It can be expressed as:

$
Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \ldots + \phi_p Y_{t-p} + \epsilon_t
$

Where:
- $ Y_t $ is the value of the time series at time $ t $.
- $ c $ is a constant term (intercept).
- $ \phi_1, \phi_2, \ldots, \phi_p $ are coefficients that represent the influence of past values.
- $ p $ is the order of the model (the number of lagged values used).
- $ \epsilon_t $ is the white noise error term, assumed to be normally distributed with a mean of 0.

### **Understanding the Parameters**

- **Order (p)**: The number of lagged observations included in the model.
- **Coefficients ($ \phi $)**: They represent how much influence each lagged observation has on the current value.
- **Error Term ($ \epsilon_t $)**: Captures randomness, unobserved factors, or measurement errors that affect $ Y_t $.

### **Example of AR(1) Model**

An AR(1) model (autoregressive model of order 1) can be written as:

$
Y_t = c + \phi_1 Y_{t-1} + \epsilon_t
$

This model implies that the current value $ Y_t $ depends only on the immediately preceding value $ Y_{t-1} $ plus a random error term.

### **Assumptions of AR Models**

1. **Linearity**: The relationship between current and past values is linear.
2. **Stationarity**: The mean, variance, and autocovariance should be constant over time.
3. **No Multicollinearity**: The lagged terms should not be highly correlated with each other.
4. **White Noise Residuals**: The error term $ \epsilon_t $ should be normally distributed with a mean of 0 and constant variance.

### **Identifying the Order (p) of AR Model**

Choosing the right order $ p $ is crucial for model accuracy. Common techniques include:

1. **Autocorrelation Function (ACF)**: Measures how values of the series are related over different time lags. If the ACF cuts off quickly after lag $ p $, it indicates an AR(p) process.
2. **Partial Autocorrelation Function (PACF)**: Shows the direct correlation between a value and its lagged values, eliminating the effects of intervening lags. The point where PACF cuts off suggests the order $ p $.

### **Fitting an AR Model**

To fit an AR model, you estimate the coefficients $ \phi_1, \phi_2, \ldots, \phi_p $. The most common method is **Ordinary Least Squares (OLS)**, where the coefficients are chosen to minimize the sum of squared differences between the actual and predicted values.

### **Characteristics of AR Models**

1. **Short-Term Prediction**: AR models are often effective for short-term forecasting, especially when past values have a strong influence on the current value.
2. **Capturing Seasonality**: Although AR models don't explicitly model seasonality, they can capture short-term seasonal effects through lagged terms.

### **Limitations of AR Models**

- **Assumes Stationarity**: AR models assume the time series is stationary, meaning you might need to difference or transform data to make it stationary.
- **Limited to Linear Relationships**: AR models capture only linear dependencies, making them less effective for non-linear time series.
- **Lag Selection**: Choosing the correct lag $ p $ is crucial; too many lags can cause overfitting, while too few can lead to underfitting.

### **Extensions of AR Models**

- **Autoregressive Moving Average (ARMA)**: Combines AR with a Moving Average (MA) model.
- **Autoregressive Integrated Moving Average (ARIMA)**: Adds integration (differencing) to handle non-stationary data.
- **Seasonal ARIMA (SARIMA)**: Extends ARIMA by incorporating seasonal components.

### **Real-World Applications of AR Models**

- **Stock Market Analysis**: Predicting future stock prices based on historical prices.
- **Economic Forecasting**: Modeling GDP, inflation rates, and other economic indicators.
- **Weather Prediction**: Short-term temperature and weather pattern forecasts.

### **Key Takeaways**

- **AR Models** are powerful for forecasting time series data, relying on past observations to predict future values.
- **Choosing the Right Order (p)** is essential, typically identified using PACF plots.
- **Stationarity** is a critical requirement; if your data isn’t stationary, transformations or differencing are needed.

The AR model is an excellent starting point for time series forecasting, providing a foundation for more complex models like ARIMA or machine learning-based approaches.



To generalize the autoregressive model of order $ p $ (AR($ p $)) into matrix form, we can express it as follows:

The AR($ p $) model is given by:
$ Y_t = c + \phi_1 Y_{t-1} + \phi_2 Y_{t-2} + \cdots + \phi_p Y_{t-p} + \epsilon_t $

If we consider all the observations from time $ t $ to $ T $, the matrix form can be written as:

$\begin{bmatrix}
Y_t \\
Y_{t+1} \\
Y_{t+2} \\
\vdots \\
Y_T
\end{bmatrix} $
$ =
\begin{bmatrix}
1 & Y_{t-1} & Y_{t-2} & \cdots & Y_{t-p} \\
1 & Y_{t} & Y_{t-1} & \cdots & Y_{t-p+1} \\
1 & Y_{t+1} & Y_t & \cdots & Y_{t-p+2} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & Y_{T-1} & Y_{T-2} & \cdots & Y_{T-p}
\end{bmatrix} $

$ \begin{bmatrix}
c \\
\phi_1 \\
\phi_2 \\
\vdots \\
\phi_p
\end{bmatrix}
+
\begin{bmatrix}
\epsilon_t \\
\epsilon_{t+1} \\
\epsilon_{t+2} \\
\vdots \\
\epsilon_T
\end{bmatrix}$

### In a compact form:
$
\mathbf{Y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\epsilon}
$

Where:
- $ \mathbf{Y} $ is the vector of observed values:
$
\mathbf{Y} = \begin{bmatrix} Y_t \\ Y_{t+1} \\ Y_{t+2} \\ \vdots \\ Y_T \end{bmatrix}
$
- $ \mathbf{X} $ is the design matrix containing a column of ones (for $ c $) and lagged values:
$
\mathbf{X} = \begin{bmatrix}
1 & Y_{t-1} & Y_{t-2} & \cdots & Y_{t-p} \\
1 & Y_{t} & Y_{t-1} & \cdots & Y_{t-p+1} \\
1 & Y_{t+1} & Y_t & \cdots & Y_{t-p+2} \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
1 & Y_{T-1} & Y_{T-2} & \cdots & Y_{T-p}
\end{bmatrix}
$
- $ \boldsymbol{\beta} $ is the parameter vector containing $ c $ and $ \phi $'s:
$
\boldsymbol{\beta} = \begin{bmatrix} c \\ \phi_1 \\ \phi_2 \\ \vdots \\ \phi_p \end{bmatrix}
$
- $ \boldsymbol{\epsilon} $ is the error vector:
$
\boldsymbol{\epsilon} = \begin{bmatrix} \epsilon_t \\ \epsilon_{t+1} \\ \epsilon_{t+2} \\ \vdots \\ \epsilon_T \end{bmatrix}
$

This matrix form provides a comprehensive general representation of any AR($ p $) model.


You can use **gradient descent** to estimate the parameters $ \phi $ and $ c $ in an autoregressive model by minimizing the **sum of squared errors (SSE)** between the predicted values and actual values of the time series. Here’s how you can approach it step-by-step, and what happens when $ \epsilon $ is sampled from a standard normal distribution.

### **Step-by-Step Gradient Descent for Autoregressive Model**

Consider the AR(1) model for simplicity:
$
Y_t = c + \phi Y_{t-1} + \epsilon_t
$
Where:
- $ Y_t $: Actual value at time $ t $
- $ c $: Intercept (constant term)
- $ \phi $: Coefficient to be learned
- $ \epsilon_t $: Error term

If you have data points $ Y_1, Y_2, \ldots, Y_T $, your goal is to find $ c $ and $ \phi $ that minimize the **mean squared error (MSE)** between the actual and predicted values.

The **MSE** can be defined as:
$
\text{MSE} = \frac{1}{T} \sum_{t=1}^{T} (Y_t - \hat{Y}_t)^2
$
Where $ \hat{Y}_t = c + \phi Y_{t-1} $.

### **Gradient Descent Algorithm**

1. **Initialize**: Start with random initial values $ c^{(0)} $ and $ \phi^{(0)} $.

2. **Compute Predictions**: For each time step $ t $, compute:
   $
   \hat{Y}_t = c^{(k)} + \phi^{(k)} Y_{t-1}
   $
   where $ k $ denotes the current iteration.

3. **Calculate Gradients**:
   The objective is to minimize:
   $
   L(c, \phi) = \frac{1}{T} \sum_{t=1}^{T} (Y_t - \hat{Y}_t)^2 = \frac{1}{T} \sum_{t=1}^{T} (Y_t - (c + \phi Y_{t-1}))^2
   $

   The gradients with respect to $ c $ and $ \phi $ are calculated as follows:

   - Gradient with respect to $ c $:
     $
     \frac{\partial L}{\partial c} = -\frac{2}{T} \sum_{t=1}^{T} (Y_t - \hat{Y}_t)
     $
     
   - Gradient with respect to $ \phi $:
     $
     \frac{\partial L}{\partial \phi} = -\frac{2}{T} \sum_{t=1}^{T} (Y_t - \hat{Y}_t) Y_{t-1}
     $

4. **Update Parameters**:
   Update $ c $ and $ \phi $ using a learning rate $ \alpha $:
   $
   c^{(k+1)} = c^{(k)} - \alpha \frac{\partial L}{\partial c}
   $
   $
   \phi^{(k+1)} = \phi^{(k)} - \alpha \frac{\partial L}{\partial \phi}
   $

5. **Repeat**: Iterate steps 2–4 until convergence (i.e., until the change in $ c $ and $ \phi $ becomes very small or until a predefined number of iterations is reached).

### **Effect of Sampling $ \epsilon_t $ from a Standard Normal Distribution**

If $ \epsilon_t $ is sampled from a standard normal distribution $ N(0, 1) $, the following properties will hold:
1. **Stationarity**: The time series $ Y_t $ is more likely to be stationary, as the errors have a mean of 0 and constant variance.
2. **Prediction Accuracy**: The model $ Y_t = c + \phi Y_{t-1} + \epsilon_t $ will capture randomness more realistically since $ \epsilon_t $ represents random shocks in real-world data. This means that even though $ Y_t $ depends on its past values, there's still an inherent randomness captured by $ \epsilon_t $.
3. **Model Fitting**: When you use gradient descent, the noise term $ \epsilon_t $ adds some variability to the updates. However, if the noise $ \epsilon_t \sim N(0, 1) $ behaves well, the gradient descent process should still converge to reasonable estimates for $ c $ and $ \phi $, provided the learning rate is appropriately chosen.

### **Why Gradient Descent Works in This Case**

The cost function $ L(c, \phi) $ is a **quadratic function** with respect to the parameters $ c $ and $ \phi $, which implies it is convex. Hence, gradient descent is well-suited for finding the global minimum.

### **Summary**

- The gradient descent method iteratively adjusts $ c $ and $ \phi $ based on the gradients of the MSE.
- Sampling $ \epsilon_t $ from a standard normal distribution represents the unpredictable fluctuations in real stock prices, making your model more robust.
- This approach works because, mathematically, the autoregressive model remains a linear regression problem in terms of past values, and gradient descent efficiently handles the optimization task.


### **OLS (Ordinary Least Square method)**:

The equation $Y = X\phi + E$ represents a fundamental concept in linear regression, particularly in the context of Ordinary Least Squares (OLS). Let's break down each component of this equation in detail:

### 1. Components of the Equation

#### a. $Y$

- **Definition**: $Y$ is the dependent variable (or response variable) that we are trying to predict or explain. It is often represented as a vector (or matrix) containing observations or measurements of the variable of interest.
- **Dimensions**: If you have $n$ observations, $Y$ is an $n \times 1$ column vector.

#### b. $X$

- **Definition**: $X$ is the design matrix (or feature matrix) that contains the independent variables (or predictors) used to predict $Y$. Each row of $X$ corresponds to an observation, and each column corresponds to a feature.
- **Dimensions**: For $n$ observations and $p$ predictors (including a column for the intercept), $X$ is an $n \times p$ matrix. The first column often consists of ones to account for the intercept term in the linear regression.

#### c. $\phi$

- **Definition**: $\phi$ is the parameter vector (or coefficient vector) that we want to estimate. Each element of $\phi$ represents the effect of the corresponding predictor on the dependent variable $Y$.
- **Dimensions**: $\phi$ is a $p \times 1$ column vector, where $p$ is the number of predictors.

#### d. $E$

- **Definition**: $E$ is the error term (or residuals), representing the difference between the actual observations and the values predicted by the linear model. It accounts for the variability in $Y$ that cannot be explained by the predictors in $X$.
- **Dimensions**: $E$ is also an $n \times 1$ column vector.

### 2. Understanding the Equation

The equation $Y = X\phi + E$ can be understood as follows:

- **Linear Relationship**: The equation states that the dependent variable $Y$ can be expressed as a linear combination of the independent variables represented in $X$, weighted by the coefficients in $\phi$, plus some error term $E$.
- **Modeling the Relationship**: In linear regression, the goal is to find the values of $\phi$ that best fit the data, meaning we want to minimize the residuals (the difference between the observed values and the values predicted by the model).

### 3. Estimation of $\phi$

To estimate the parameters $\phi$, we typically use the Ordinary Least Squares (OLS) method, which minimizes the sum of squared residuals. Here's a brief outline of the OLS estimation process in the context of this equation:

1. **Define Residuals**:
   - The residuals can be defined as $E = Y - X\phi$.

2. **Objective Function**:
   - The objective is to minimize the Residual Sum of Squares (RSS):
   $
   \text{RSS} = \|Y - X\phi\|^2
   $

3. **Differentiation**:
   - By taking the derivative of the RSS with respect to $\phi$ and setting it to zero, you obtain the normal equations:
   $
   X^\top Y = X^\top X \phi
   $

4. **Solve for $\phi$**:
   - The OLS solution for $\phi$ can be found using:
   $
   \phi = (X^\top X)^{-1} X^\top Y
   $
   - This provides the estimated coefficients that minimize the residuals.

### 4. Assumptions in Linear Regression

For the OLS estimates of $\phi$ to be valid, several assumptions must hold:

1. **Linearity**: The relationship between the independent and dependent variables is linear.
2. **Independence**: The residuals $E$ are independent of each other.
3. **Homoscedasticity**: The residuals have constant variance across all levels of the independent variables.
4. **Normality**: The residuals are normally distributed (this is especially important for hypothesis testing).

### Summary

The equation $Y = X\phi + E$ captures the essence of linear regression, where the goal is to find the best linear approximation of $Y$ using the predictors in $X$ through the estimation of parameters $\phi$.
```python
def load_stock_data(company_name, file_extension="csv"):
    # Updated path with flexibility for Excel or CSV files
    file_path = f"./data/{company_name}.{file_extension}"

    # Check if the file exists
    if os.path.exists(file_path):
        print(f"File found at: {file_path}")
    else:
        print(f"File not found at: {file_path}")
        return None

    # Load the file based on the extension
    try:
        if file_extension == "xlsx":
            df = pd.read_excel(file_path)  # Load Excel file
        elif file_extension == "csv":
            df = pd.read_csv(file_path)  # Load CSV file
        else:
            print("Unsupported file extension.")
            return None
        
        # Convert "Month" to datetime format (e.g., 'Sep-14' -> '2014-09')
        df['Month'] = pd.to_datetime(df['Month'], format='%B-%Y')
        return df
    except Exception as e:
        print(f"Error loading file: {e}")
        return None
```

```python
def get_stock_price(company_name, year, month, file_extension="csv"):
    df = load_stock_data(company_name, file_extension)
    if df is None:
        return None

    # Filter the data by year and month
    filtered_df = df[(df['Month'].dt.year == year) & (df['Month'].dt.strftime('%B') == month)]
    return filtered_df[['Month', 'Close Price']]  # Display only the month and "Close Price"
```

```python
def compare_stocks(company1, company2, year, month, file_extension="csv"):
    stock1 = get_stock_price(company1, year, month, file_extension)
    stock2 = get_stock_price(company2, year, month, file_extension)
    if stock1 is None or stock2 is None:
        return None
    
    comparison = pd.DataFrame({
        'Company 1': stock1['Close Price'].values,
        'Company 2': stock2['Close Price'].values
    })
    
    return comparison
```

```python
def predict_stock(company_name, file_extension="csv"):
    df = load_stock_data(company_name, file_extension)
    if df is None:
        return None

    # Compute historical trend (e.g., 5-year moving average)
    df['Moving_Avg'] = df['Close Price'].rolling(window=60).mean()  # 5 years = 60 months
    
    # Assuming the future follows the same trend, extend the moving average for the next 5 years
    future_dates = pd.date_range(start='2024-10', periods=60, freq='M')
    future_pred = np.linspace(df['Close Price'].mean(), df['Close Price'].mean() * 1.1, 60)  # Dummy linear increase

    prediction_df = pd.DataFrame({
        'Month': future_dates.strftime('%b %y'),
        'Predicted_Close': future_pred
    })

    return prediction_df
```

```python
def calculate_risk(company_name, file_extension="csv"):
    df = load_stock_data(company_name, file_extension)
    if df is None:
        return None

    # Calculate monthly returns
    df['Returns'] = df['Close Price'].pct_change()
    volatility = df['Returns'].std()  # Standard deviation of returns as a measure of risk
    return volatility * 100  # Risk percentage
```

```python

class AutoRegression:
    def __init__(self, data, end_point): # data: dataframe ,Hdfc 
        """
        Initialize the AutoRegression model with data and the prediction end date.

        Parameters:
        - data: A pandas DataFrame containing the time series data.
        - end_point: The date up to which predictions are to be made in 'Month-Year' format.
        """
        # Ensure 'Month' is string and 'Year' is integer
        data[['Month', 'Year']] = data['Month'].str.split('-', expand=True)

        # Convert the 'Year' column to integer
        data['Year'] = data['Year'].astype(int)

        # Convert the 'Month-Year' format to datetime
        data['Date'] = pd.to_datetime(data['Month'] + '-' + data['Year'].astype(str), format='%B-%Y')
        self.data = data.set_index('Date')  # Set the Date as index
        self.end_point = pd.to_datetime(end_point, format='%B-%Y')  # Convert end_point to a datetime object
        self.predictions = None  # Placeholder for predictions
        self.phi = None  # Placeholder for phi coefficients

    def fit(self, column, p):
        """
        Fit the autoregressive model.

        Parameters:
        - column: The name of the column to predict.
        - p: The number of lag observations (autoregressive terms).
        """
        self.column = column
        self.p = p
        self.y = self.data[column].values

        # Create lagged values for the design matrix
        X = np.array([self.y[i-p:i] for i in range(p, len(self.y))])
        self.y_target = self.y[p:]

        # Add a constant term for the intercept in OLS
        X_with_intercept = np.column_stack((np.ones(X.shape[0]), X))

        # Calculate the coefficients (phi) using Ordinary Least Squares
        self.phi = np.linalg.lstsq(X_with_intercept, self.y_target, rcond=None)[0]

    def predict(self):
      """
      Make predictions until the specified end date.
      Returns the predictions as a DataFrame.
      """
      # Get the last p observations for predictions
      last_values = self.y[-self.p:]

      predictions = []
      current_date = self.data.index[-1]  # Start from the last date in the data

      while current_date < self.end_point:
          # Calculate the next prediction using the linear combination of the last p values and phi coefficients
          next_prediction = self.phi[0] + np.sum(self.phi[1:] * last_values)
          predictions.append(next_prediction)

          # Update last_values with the new prediction
          last_values = np.roll(last_values, -1)  # Shift values to the left
          last_values[-1] = next_prediction  # Add the new prediction at the end

          # Update the date to the next month (skip day)
          current_date += pd.DateOffset(months=1)
          current_date = current_date.replace(day=1)  # Ensure the day is set to 1

         # Create a DataFrame for the predictions with corresponding dates
      prediction_index = pd.date_range(start=self.data.index[-1] + pd.DateOffset(months=1),
                                      periods=len(predictions), freq='M')
      self.predictions = pd.DataFrame({
        'DATE': prediction_index.strftime('%B-%Y'),  # Format as 'Month-Year'
        self.column: predictions
      })

      return self.predictions

```
